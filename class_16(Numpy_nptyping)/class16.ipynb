# Numpy with NDArray typing support

````python
from nptyping import NDArray,Shape,UInt64
from typing import Any

data : NDArray[Shape["10"],Any] = np.arange(1,11)
print(data)
print(data + 5)         
````

# numpy advanced operations vs list operations

````python
data : list[int] = list(range(1,21))
print(data)
print(data[5:11])
data[5:11] = 1000 # normal python list
print(data)
````

# Numpy 1d functions

````python
from nptyping import Bool

state_bank : NDArray[Shape["10"],Any] = np.array([1,7,8,10])

copetitor : NDArray[Shape["10"],Bool] = np.array([True,False,False,True])

state_bank[copetitor]
````

````python
state_bank : NDArray[Shape["4"],Any] = np.array([1,7,8,10])
state_bank%2==0
````

````python
state_bank : NDArray[Shape["4"],Any] = np.array([1,7,8,10])
state_bank[state_bank%2==0]
````

````python
state_bank : NDArray[Shape["10"],Any] = np.array([1,7,8,10])
ubl_bank : NDArray[Shape["100"],Any] = np.random.randint(1,22,3)
mzn_bank : NDArray[Shape["100"],Any] = np.random.randint(1,22,20)
display(state_bank)
display(ubl_bank)
display(mzn_bank)
````

```
in1d is a function in the NumPy library in Python. It is used to test whether each element of a given array is present in another array. The function returns a boolean array that has the same length as the first array, where each value is True if the element is in the second array and False otherwise.
```

````python
np.in1d(state_bank,ubl_bank) # only 7 and 8 were found in match/intersection
````

````python
state_bank[np.in1d(state_bank,ubl_bank)]
````

````python
np.intersect1d(state_bank,ubl_bank) # will print the intersection elements
````

````python
group1 : NDArray[Shape["10"],Any] = np.array([1,3,4,5,7])
group2 : NDArray[Shape["10"],Any] = np.array([6,3,2,5,100])

display(group1,group2)
np.where(group1>group2, group1,group2)
````

# Create Numpy arrays

````python
a : NDArray[Shape["Size, Size"], Any] = np.array([[1, 2, 3],
                                                      [4, 5, 6]])

b : NDArray[Shape["Size , Size"],Any] = np.array([[2,6],
                                                [7,8]])

c : NDArray[Shape["Size , Size"],Any] = np.array([["A"],
                                                  ["B"]])

print(a)
print(b)
print(c)
````

````python
x : NDArray[Shape["* , *"],Any] = np.array([[1,2,3],[3,4,5]])
y : NDArray[Shape["* , *"],Any] = np.array([[4,3],[1,1]])
n : NDArray[Shape["* , *"],Any] = np.array([["s"],["d"]])

print(x)
print(y)
print(n)
````

# Create any dimension array
